# Problem Statement :
The following data contains the attributes contributing to the dissolved oxygen level in River water. Create a model to predict the level of Oxygen dissolved in the River water


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, RandomizedSearchCV, cross_val_score

from sklearn.linear_model import LinearRegression, Ridge

from xgboost import XGBClassifier
import xgboost as xg
from sklearn.metrics import mean_squared_error as MSE

from sklearn.ensemble import RandomForestRegressor
```

```python
#Reading csv file
data = pd.read_csv('train.csv')
```

```python
df = data.copy()
df.head()
```

```python
#checking no of rows and columns 
df.shape
```

# Preprocessing 

```python
#drop Id column as it is unique identifier
df.drop(['Id'], axis=1, inplace=True)
df
```

```python
#checking if there is any feature of object type
df.info()
```

```python
#checking the sum of null values for each feature.. 
df.isnull().sum()
```

```python
#dropping features with more than 60% nan values
#60% of 147 rows is 88
147*60/100
```

```python
for col in df.columns:
    if df[col].isnull().sum() > 88:
        df.drop([col], axis = 1, inplace=True)
        
df.shape   
```

```python
df.head()
```

# Exploratory data analysis

```python
#Correlation Heatmap

corr = df.corr()
plt.figure(figsize=(8,8))
sns.heatmap(corr, annot=True, cmap='RdYlGn')
```

Correlation features

1. NH4_1 and NH4_2 have 84% correlation - we will drop NH4_2 as it is less correlated with target(0.11)
2. BOD5_1 and BOD5_2 are 74% correlated - we will drop BOD5_2 as it is less correlated with target (0.084)

```python
df.drop(['NH4_2', 'BOD5_2'], axis=1, inplace=True)
df.head()
```

# Filling NAN with mean values

```python
df.isnull().sum()
```

```python
for col in df.columns:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].mean(), inplace=True)
        
```

```python
df.isnull().sum()
```

# Regression 

Train test split

```python
X = df.copy()
X
```

```python
X.drop(['target'], axis=1, inplace=True)
X.head()
```

```python
y = df['target']
```

```python
y.head()
```

```python
X_train,X_test, y_train, y_test = train_test_split( X, y, test_size=0.33, random_state=42 )
```

```python
regressor = RandomForestRegressor(n_estimators = 100, random_state = 0)
regressor.fit(X_train, y_train)
regressor.score(X_test, y_test)
```

# Feature scaling

```python
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
X_train
```

```python
X_train = pd.DataFrame(X_train,columns=X.columns)
X_train.head()
```

```python
X_test = pd.DataFrame(X_test, columns=X.columns)
X_test.head()
```

# Training Model

Linear regression

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```

```python
ridge = Ridge()
ridge.fit(X_train, y_train)
ridge.score(X_test, y_test)
```

```python
#xgb_r = xg.XGBRegressor(objective ='reg:linear',n_estimators = 10, seed = 123)
#xgb_r.fit(X_train, y_train)
#xgb_r.score(X_test, y_test)
regressor = RandomForestRegressor(n_estimators = 100, random_state = 0)
regressor.fit(X_train, y_train)
regressor.score(X_test, y_test)
```

Predicting exact values does not give us that good results as there is not enough correlation between the I/p & O/p features to predict the exact values of the O/p.
So we try predicting whether Oxygen Level in River water is either High or Low
Part 2
Classification (Predicting High or Low)

```python
y_new = pd.qcut(y,q=2, labels=[0,1]) # less than mean 0 .. more 1
```

```python
y_new.head()
```

```python
X.head()
```

# Training 

```python
xgboost = XGBClassifier()
```

# Hyper params tuning

```python
params={
 "learning_rate"    : [0.05, 0.10, 0.15, 0.20, 0.25, 0.30 ] ,
 "max_depth"        : [ 3, 4, 5, 6, 8, 10, 12, 15],
 "min_child_weight" : [ 1, 3, 5, 7 ],
 "gamma"            : [ 0.0, 0.1, 0.2 , 0.3, 0.4 ],
 "colsample_bytree" : [ 0.3, 0.4, 0.5 , 0.7 ]
    
}
```

```python

rs_cv = RandomizedSearchCV(xgboost, param_distributions=params, n_iter=5, n_jobs=1, cv=5)
```

```python
rs_cv.fit(X, y_new)
```

```python
score = cross_val_score(xgboost, X, y_new, cv=10)
```

```python
rs_cv.best_params_
```
